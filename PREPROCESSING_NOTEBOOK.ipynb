### 1.4 Optmizing datatypes to save space

```python
patient_data['Visit Date'] = pd.to_datetime(patient_data['Visit Date'])

category_cols = [
    'Gender', 'City of Residence', 'Profession',
       'Insurance Provider', 'Family History', 'Education Level',
       'Marital Status', 'Department'
]
patient_data[category_cols] = patient_data[category_cols].astype("category")

patient_data['Consultation Duration'] = pd.to_timedelta(patient_data['Consultation Duration'])
patient_data['Satisfaction Level'] = patient_data['Satisfaction Level'].astype("int8")
patient_data['Approximate Annual Income'] = patient_data['Approximate Annual Income'].astype("float32")
patient_data['Consultation Price'] = patient_data['Consultation Price'].astype("float16")
patient_data['Insurance Coverage'] = patient_data['Insurance Coverage'].astype("float16")

patient_data.info()
```

#### Patient specific and visit level features

To handle missing values appropriately, we separated the features two lists. Some variables represent **stable patient characteristics**, while others correspond to **individual visit details**.  



```python
# Patient specific
pat_cols = [
    'Age','Insurance Provider', 'Family History', 'Education Level','Approximate Annual Income'
]

# Visit-level numeric fields
visit_num = ['Consultation Duration','Consultation Price','Insurance Coverage']
```

```python

# Fill within each Patient ID by propagating known values
patient_data[pat_cols] = patient_data.groupby('Patient ID')[pat_cols].transform(
    lambda x: x.ffill().bfill())
```

```python
for col in visit_num:
    patient_data[col] = patient_data.groupby('Department')[col].transform(
        lambda x: x.fillna(x.median()))
```

```python
patient_data.isnull().sum()
```

```python
print(patient_data['Insurance Provider'].value_counts())
print('--------')
print(patient_data['Family History'].value_counts())
```

```python
# Add new categories first before filling
patient_data['Insurance Provider'] = patient_data['Insurance Provider'].cat.add_categories(['No Provider'])
patient_data['Family History'] = patient_data['Family History'].cat.add_categories(['No Disease'])

# Now fill with the new categories
patient_data['Insurance Provider'].fillna('No Provider', inplace=True)
patient_data['Family History'].fillna('No Disease', inplace=True)

# Verify no missing values remain
patient_data.isnull().sum()
```
